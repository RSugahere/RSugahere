5. Performing visualisations.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

5.1. Comparing Width of each Flower Category

```python
sns.set_style('darkgrid')
plt.bar(df['Classification'],df['SW'])
plt.bar(df['Classification'],df['PW'])
plt.title('Width of each classification',size=20,c='g')
plt.xlabel('Classification',size=15,c='g')
plt.ylabel('Width',size=15,c='g')
plt.legend(['SW','PW'])
plt.show()
```

5.2. Comparing Length of each Flower Category

```python
sns.set_style('darkgrid')
plt.bar(df['Classification'],df['SL'])
plt.bar(df['Classification'],df['PL'])
plt.title('Length of each classification',size=20,c='g')
plt.xlabel('Classification',size=15,c='g')
plt.ylabel('Length',size=15,c='g')
plt.legend(['SL','PL'])
plt.show()
```

5.3. Finding the distribution of Width(SW & PW) and Length(SL & PL)

```python
plt.subplot(2,2,1)
plt.boxplot(df['SW'])
plt.ylabel('Sepal Width',size=10,c='g')
plt.subplot(2,2,2)
plt.boxplot(df['PW'])
plt.ylabel('Petal Width',size=10,c='g')
plt.subplot(2,2,3)
plt.boxplot(df['SL'])
plt.ylabel('Sepal Length',size=10,c='g')
plt.subplot(2,2,4)
plt.boxplot(df['PL'])
plt.ylabel('Petal Length',size=10,c='g')
```

5.4. Comparing Sepal attributes of each category

```python
plt.title('Length vs Width (Sepal)',size=23,c='g')
sns.scatterplot(df['SL'],df['SW'],hue=df['Classification'],c='b',marker='D')
plt.xlabel('SL',size=15,c='g')
plt.ylabel('SW',size=15,c='g')
plt.show()
```

5.5. Comparing Petal attributes of each category

```python
plt.title('Length vs Width (Petal)',size=23,c='g')
sns.scatterplot(df['PL'],df['PW'],hue=df['Classification'],c='b',marker='D')
plt.xlabel('PL',size=15,c='g')
plt.ylabel('PW',size=15,c='g')
plt.show()
```
